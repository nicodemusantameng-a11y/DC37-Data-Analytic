**Another Visualization**

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your specific file
df = pd.read_csv('/content/NYPD_Arrest_Data_(Year_to_Date)_20260411.csv')

# Preprocessing: Map Borough codes to full names for readability
boro_map = {'B': 'Bronx', 'S': 'Staten Island', 'K': 'Brooklyn', 'M': 'Manhattan', 'Q': 'Queens'}
df['BOROUGH_NAME'] = df['ARREST_BORO'].map(boro_map)

# Set a "fancy" style with a grey background and clean gridlines
sns.set_theme(style="whitegrid")
plt.rcParams['font.family'] = 'sans-serif'

# Create a 1-row, 2-column dashboard
fig, ax = plt.subplots(1, 2, figsize=(18, 8))
fig.patch.set_facecolor('#f7f7f7')

# Panel 1: Heatmap of Intensity
pivot_df = df.groupby(['BOROUGH_NAME', 'LAW_CAT_CD']).size().unstack(fill_value=0)
sns.heatmap(pivot_df, annot=True, fmt="d", cmap="YlGnBu", ax=ax[0], cbar_kws={'label': 'Count'})
ax[0].set_title('Arrest Intensity by Borough & Level', fontsize=16, fontweight='bold', pad=20)
ax[0].set_xlabel('Offense Severity', fontsize=12)
ax[0].set_ylabel('Borough', fontsize=12)

# Panel 2: Demographic Breakdown
# Standardize common NYPD age labels for sorting
age_order = ['<18', '18-24', '25-44', '45-64', '65+']
df['AGE_GROUP'] = pd.Categorical(df['AGE_GROUP'], categories=age_order, ordered=True)

sns.countplot(data=df, x='BOROUGH_NAME', hue='AGE_GROUP', palette='magma', ax=ax[1])
ax[1].set_title('Arrestee Age Distribution by Borough', fontsize=16, fontweight='bold', pad=20)
ax[1].set_xlabel('Borough', fontsize=12)
ax[1].set_ylabel('Total Arrests', fontsize=12)
ax[1].legend(title='Age Group', loc='upper right')

plt.tight_layout()
plt.show()
```
